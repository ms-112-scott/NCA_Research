```python
import numpy as np
```

```python
def inspect_obj(obj, indent=0, max_depth=3):
    """遞迴列印物件結構"""
    prefix = "  " * indent
    t = type(obj)
    
    if isinstance(obj, np.ndarray):
        print(f"{prefix}- type: np.ndarray, shape={obj.shape}, dtype={obj.dtype}")
    elif isinstance(obj, dict):
        print(f"{prefix}- type: dict, len={len(obj)}")
        if indent < max_depth:
            for k, v in obj.items():
                print(f"{prefix}  key: {k}")
                inspect_obj(v, indent + 2, max_depth)
    elif isinstance(obj, (list, tuple)):
        print(f"{prefix}- type: {t.__name__}, len={len(obj)}")
        if indent < max_depth and len(obj) > 0:
            print(f"{prefix}  [0]:")
            inspect_obj(obj[0], indent + 2, max_depth)
    else:
        try:
            print(f"{prefix}- type: {t.__name__}, len={len(obj)}")
        except Exception:
            print(f"{prefix}- type: {t.__name__}, value={repr(obj)[:80]}")
```

```python

path = r"C:\Users\GAI\Desktop\Scott\NCA_Research\E00_lstm\dataset\LFMG_01.npz"
data = np.load(path, allow_pickle=True)

print(f"Loaded NPZ file: {path}")
print(f"Keys: {data.files}\n")

for key in data.files:
    print(f"Key: {key}")
    obj = data[key]
    inspect_obj(obj)
    print()
```
