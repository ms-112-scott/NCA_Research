```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_all_png(root_dir):
    png_files = []
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            if file.lower().endswith(".png") and "E4-4.3" in dirpath:
                png_files.append(os.path.join(dirpath, file))

    print(f"找到 {len(png_files)} 張圖片")

    for i, f in enumerate(png_files):
        img = mpimg.imread(f)
        rel_path = os.path.relpath(f, root_dir)  # 相對路徑
        plt.imshow(img)
        plt.title(f"{i+1}/{len(png_files)}: {rel_path}")
        plt.axis("off")
        plt.show()

# 使用範例
show_all_png(r"C:\Users\GAI\Desktop\Scott\NCA_Research\E4_PI_NCA\outputs")
```
